#### Feature engineering

```python
###Date
df['arrival_date']=pd.to_datetime( df["arrival_date_year"].astype(str) + "-" +
    df["arrival_date_month"] + "-" +
    df["arrival_date_day_of_month"].astype(str)
                                 )
```

```python
df['arrival_date'].dtype
```

```python
df['total_guests']=df['adults']+df['children']+df['babies']
df['total_guests']
```

```python
df["total_nights"] = df["stays_in_weekend_nights"] + df["stays_in_week_nights"]
df['total_nights']
```

```python
##know if booking is  MADE BY a Family or not
df["is_family"] = ((df["children"] > 0) | (df["babies"] > 0)).astype(int)
df['is_family']
```

```python
df['estimated_revenue']=df['adr']*df['total_nights']
df['estimated_revenue']
```

```python
df["lead_category"] = pd.cut(
    df["lead_time"],
    bins=[-1, 7, 30, 90, 365],
    labels=["Very Short", "Short", "Medium", "Long"]
)
df['lead_category']     ###Long lead bookings cancel more
                        ###Short lead bookings cancel less
```

```python
df["is_weekend_stay"] = (df["stays_in_weekend_nights"] > 0).astype(int)
df['is_weekend_stay']
```

```python
df["is_weekend_stay"] = (df["stays_in_weekend_nights"] > 0).astype(int)
df['is_weekend_stay']
```

```python
df["high_request"] = (df["total_of_special_requests"] >= 2).astype(int)
df['high_request']
```

```python
###High risk guest
df["cancellation_rate_history"] = (
    df["previous_cancellations"] /
    (df["previous_cancellations"] + df["previous_bookings_not_canceled"] + 1)
)
df['cancellation_rate_history'].tail(1000)  ### 0.0 -> first time guests
```

```python
df['is_repeated_guest'].value_counts()
```

```python
##Seasonality
df['arrival_month']=df['arrival_date'].dt.month
df['arrival_month']
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

df["season"] = df["arrival_month"].apply(get_season)
df['season']
```

```python
##Agent vs direct booking
df["is_direct_booking"] = (df['is_agent_booking'] == 0).astype(int)
df['is_direct_booking']
```

```python
##corporate booking flag
df["is_corporate"] = (df["company"] != 0).astype(int)
df['is_corporate']
```

```python
##booking changes
df["has_booking_changes"] = (df["booking_changes"] > 0).astype(int)
```

```python
df['has_booking_changes']
```

```python
df["non_refundable"] = (df["deposit_type"] == "Non Refund").astype(int)
df['non_refundable']
```

#### KEY FINDINGS:

###### Cancellation rate is high and impacts revenue.

###### Long lead-time bookings cancel more frequently.

###### Customers with previous cancellations are high-risk.

###### O TA bookings contribute heavily to cancellations.

###### High ADR bookings cause higher revenue loss when canceled.

#### TOP RISK SEGMENTS:

###### Long lead-time bookings

###### Online Travel Agent bookings

###### High ADR reservations

###### Customers with past cancellations

#### SOLUTIONS:

###### Introduce partial deposit for high-risk bookings.

###### Send confirmation reminders to repeat cancellers.

###### Review OTA cancellation policies.
